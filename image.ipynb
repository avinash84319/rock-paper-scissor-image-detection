```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from tensorflow.keras.models import load_model
model=load_model('rps.h5')
```

```python
import time
```

```python
cam=cv2.VideoCapture(0)
while True:
    ret,frame=cam.read()
    frame=cv2.flip(frame,1)
    image=cv2.resize(frame,(150,150))
    pred=model.predict(np.array([image]))
    pred=np.argmax(pred)
    if pred==0:
        cv2.putText(frame,'Paper',(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
    elif pred==1:
        cv2.putText(frame,'Rock',(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
    elif pred==2:
        cv2.putText(frame,'Scissor',(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
    if cv2.waitKey(20) & 0xFF == ord('d'):
        break
    cv2.imshow('frame',frame)
cam.release()
cv2.destroyAllWindows()
```
